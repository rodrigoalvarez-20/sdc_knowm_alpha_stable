```python
import numpy as np
import matplotlib.pyplot as plt
from memristor_hp import HPMemristor
```

```python
# Par치metros de simulaci칩n
#dt = 1e-6  # Paso de tiempo en segundos
t_max = 0.5  # Tiempo m치ximo de simulaci칩n en segundos
timesteps = 1000 #int(t_max / dt)
frequency = 10  # Hz
amplitude = 1  # Voltio
time = np.linspace(0, t_max, timesteps)
V = amplitude * np.sin(1 * np.pi * frequency * time)
```

```python
plt.plot(V)
```

```python
hp_memres = HPMemristor()
```

```python
hp_memres.set_input_voltage(V)
```

```python
hp_memres.run_simulation()
```
